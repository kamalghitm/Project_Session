```python
x=df.drop(df.columns[0],axis=1)
x
```

```python
y=df.iloc[:,0]
y.shape
```

```python
sc=StandardScaler();
x=sc.fit_transform(x)
x
```

```python
parameters={'criterion':['gini','entropy']}
gd_cv=GridSearchCV(DecisionTreeClassifier(),parameters,cv=10)
gd_cv.fit(x,y)
print("Best param for DecisionTreeClassifier : ",gd_cv.best_params_)

cv_score=cross_val_score(DecisionTreeClassifier(criterion='gini'),x,y,cv=10)
print(cv_score.mean())

```

```python
parameters={'n_neighbors':[5,10,15,20,30]}
gd_cv=GridSearchCV(KNeighborsClassifier(),parameters,cv=10)
gd_cv.fit(x,y)
print("Best param for KNeighborsClassifier : ",gd_cv.best_params_)

cv_score=cross_val_score(KNeighborsClassifier(n_neighbors=5),x,y,cv=10)
print(cv_score.mean())
```

```python
parameters={'kernel':['rbf','linear','poly']}
gd_cv=GridSearchCV(SVC(),parameters,cv=10)
gd_cv.fit(x,y)
print("Best param for SVC : ",gd_cv.best_params_)

cv_score=cross_val_score(SVC(kernel='poly'),x,y,cv=10)
print(cv_score.mean())
```

```python
#good for binary classification
cv_score=cross_val_score(LogisticRegression(),x,y,cv=10)
print(cv_score.mean())

cv_score=cross_val_score(GaussianNB(),x,y,cv=10)
print(cv_score.mean())
```

```python
parameters={'n_estimators':[5,10,50,100],'criterion':['gini','entropy']}
gd_cv=GridSearchCV(RandomForestClassifier(),parameters,cv=10)
gd_cv.fit(x,y)
print("RandomForestClassifier best_params_",gd_cv.best_params_)

cv_score=cross_val_score(RandomForestClassifier(n_estimators=5,criterion='gini'),x,y,cv=10)
print(cv_score.mean())
```

```python
from sklearn.metrics import r2_score
max_rand_state=0
max_r2_score=0
r_score=0
for ran_state in range(42,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=ran_state)
    #svc=SVC(kernel='poly')
    #svc=RandomForestClassifier(n_estimators=5,criterion='entropy')
    svc=LogisticRegression()
    svc.fit(x_train,y_train)
    y_pred=svc.predict(x_test)
    r_score=r2_score(y_test,y_pred)
    if(r_score>max_r2_score):
        max_r2_score=r_score
        max_rand_state=ran_state
    
   
    

print("max random state value :",max_rand_state," and corresponding max r2_score :",max_r2_score)    
```

```python
#we can find overfitting isssue by checking score of train data and test data .If train data score more than test data then it is overfitting
#for classification check score like accuracy_score,mean_squared_error,mean_absolute_error,RMSE(sqrt of MSE),confusion_matrix,classification_report
#for regression check r2_score,mean_squared_error,mean_absolute_error,RMSE(sqrt of MSE)

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=96)
#svc=SVC(kernel='poly')
#svc=RandomForestClassifier(n_estimators=5,criterion='entropy')
svc=LogisticRegression()
svc.fit(x_train,y_train)
y_test_pred=svc.predict(x_train)
print(r2_score(y_train,y_test_pred),mean_squared_error(y_train,y_test_pred),mean_absolute_error(y_train,y_test_pred),accuracy_score(y_train,y_test_pred),confusion_matrix(y_train,y_test_pred),classification_report(y_train,y_test_pred))
y_pred=svc.predict(x_test)

print(r2_score(y_test,y_pred),mean_squared_error(y_test,y_pred),mean_absolute_error(y_test,y_pred),accuracy_score(y_test,y_pred),confusion_matrix(y_test,y_pred),classification_report(y_test,y_pred))
```
